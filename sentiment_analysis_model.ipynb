```
def levenstein(word,given_list,lev_index,s_index):
  elist = []
  for i in given_list:
    if word[0] ==i[0]:
      elist.append(i)
    else:
      pass

  pair = {}
  for i in elist:
    lev_dist = edit_distance(word,i)
    if lev_dist<=lev_index:
      pair.update({i:lev_dist})
    
    
  
  if not bool(pair):
    return 42
    
  


  pair2  = {}

  for key, value in sorted(pair.items(), key=lambda item: item[1]):
      pair2.update({key:value})

    
  pair2_list = []
  count2 = 0
  for key in pair2.keys() :
    count2 = count2 +1
    if count2<20:
      
      pair2_list.append(key)
      
        
      
    else:
      pass
  
  
  if not pair2_list:
    return 42

  
  def soundex(word):
     codes = ("bfpv","cgjkqsxz", "dt", "l", "mn", "r")
     soundDict = dict((ch, str(ix+1)) for ix,cod in enumerate(codes) for ch in cod)
     cmap2 = lambda kar: soundDict.get(kar, '9')
     sdx =  ''.join(cmap2(kar) for kar in word.lower())
     sdx2 = word[0].upper() + ''.join(k for k,g in list(groupby(sdx))[1:] if k!='9')
     sdx3 = sdx2[0:4].ljust(4,'0')
     return sdx3


  pair2_list_soundex = {}
  for i in pair2_list:
    pair2_list_soundex.update({i:soundex(i)})

  a = soundex(word)
  a =int(a[1:])

  new_dict = {}
  for i in pair2_list_soundex:
    b=str(pair2_list_soundex.get(i))
    b=int(b[1:])
    new_dict.update({i:b})

  nd = {}
  for i in new_dict:
    if abs(a-new_dict.get(i))<=s_index:
      nd.update({i:abs(a-new_dict.get(i))})
    
    
  if not nd:
    return 42
       
        
         
    

  sorted_x = sorted(nd.items(), key=lambda kv: kv[1])
  import collections
  sorted_dict = collections.OrderedDict(sorted_x)
  lev_word = list(sorted_dict.keys())[0]
 
  return lev_word
```

```
wordfreq = {}  
for sentence in corpus_copy2:  
    tokens = nltk.word_tokenize(sentence)
    for token in tokens:
        if token not in wordfreq.keys():
            wordfreq[token] = 1
        else:
            wordfreq[token] += 1
```

```
len(wordfreq)
```

```
import heapq  
most_freq = heapq.nlargest(2000, wordfreq, key=wordfreq.get)
```

```
sentence_vectors = []  
for sentence in corpus_copy2:  
    sentence_tokens = nltk.word_tokenize(sentence)
    sent_vec = []
    for token in most_freq:
        if token in sentence_tokens:
            sent_vec.append(1)
        else:
            sent_vec.append(0)
    sentence_vectors.append(sent_vec)
```

```

def premat(cm):
  accuracy = (cm[0][0]+cm[1][1])/(cm[0][0]+cm[1][1]+cm[0][1]+cm[1][0])
  print("accuracy", accuracy)
  precision = cm[1][1]/(cm[1][1]+cm[0][1])
  print("precision", precision)
  recall = cm[1][1]/(cm[1][1]+cm[1][0])
  print("recall",recall)
  f1 = 2*precision*recall/(precision+recall)
  print("f1 score ",f1)
  
premat(cm)
```
